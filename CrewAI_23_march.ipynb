```python
import os
from crewai import Agent, Task, Crew
from google.colab import userdata
from langchain_litellm import ChatLiteLLM

# =====================================
# 🔑 1. Load Gemini API Key (SECURE)
# =====================================

api_key = userdata.get("faltu_kaam1")

if not api_key:
    raise ValueError("❌ GEMINI_API_KEY not found in Colab Secrets")

os.environ["faltu_kaam1"] =api_key

# Remove other keys
for key in ["OPENAI_API_KEY", "GROQ_API_KEY"]:
    if key in os.environ:
        del os.environ[key]

print("✅ API Key Loaded Successfully")

# =====================================
# 🤖 2. Define LLM (Gemini)
# =====================================
llm = ChatLiteLLM(
    model="gemini/gemini-2.0-flash-exp"
)

# =====================================
# 👨‍💻 3. Define Agents
# =====================================

trend_agent = Agent(
    role="Trend Monitoring Agent",
    goal="Find latest EdTech innovations and AI in education trends",
    backstory="Expert in tracking global EdTech developments and innovations",
    llm=llm,
    verbose=True
)

summary_agent = Agent(
    role="Content Summarization Agent",
    goal="Summarize educational trends into simple insights",
    backstory="Skilled in simplifying complex educational concepts for learners",
    llm=llm,
    verbose=True
)

writer_agent = Agent(
    role="Blog Writing Agent",
    goal="Write engaging and SEO-friendly blog posts",
    backstory="Professional blog writer specializing in EdTech content",
    llm=llm,
    verbose=True
)

publisher_agent = Agent(
    role="Publishing Agent",
    goal="Prepare and schedule blog posts for publishing",
    backstory="Responsible for formatting and publishing content online",
    llm=llm,
    verbose=True
)

# =====================================
# 📋 4. Define Tasks
# =====================================

trend_task = Task(
    description="Find 5 latest trends in EdTech and AI in education (2026)",
    expected_output="List of trends with explanation",
    agent=trend_agent
)

summary_task = Task(
    description="Summarize the trends into simple insights for students and educators",
    expected_output="Short summarized insights",
    agent=summary_agent,
    context=[trend_task]
)

blog_task = Task(
    description="Write a weekly blog post based on the summarized insights",
    expected_output="Well-structured blog post with title, intro, body, and conclusion",
    agent=writer_agent,
    context=[summary_task]
)

publish_task = Task(
    description="Format the blog and prepare it for publishing with headings and SEO keywords",
    expected_output="Final publish-ready blog content",
    agent=publisher_agent,
    context=[blog_task]
)

# =====================================
# 👥 5. Create Crew (Multi-Agent System)
# =====================================

crew = Crew(
    agents=[trend_agent, summary_agent, writer_agent, publisher_agent],
    tasks=[trend_task, summary_task, blog_task, publish_task],
    verbose=True
)

# =====================================
# 🚀 6. Run System
# =====================================

try:
    result = crew.kickoff()

    print("\n🎉 FINAL BLOG OUTPUT:\n")
    print(result)

except Exception as e:
    print("❌ ERROR OCCURRED:")
    print(str(e))
```

```python

import os
from google import genai
from google.colab import userdata

# =====================================
# 🔑 Load API Key
# =====================================
api_key = userdata.get("faltu_kaam3")

if not api_key:
    raise ValueError("❌ GEMINI_API_KEY not found")

client = genai.Client(api_key=api_key)

print("✅ New Gemini API Ready")

# =====================================
# 🤖 Gemini Call Function
# =====================================

def gemini_call(prompt):
    response = client.models.generate_content(
        model="gemini-2.0-flash",   # ✅ WORKING MODEL
        contents=prompt
    )
    return response.text

# =====================================
# 👨‍💻 Agent Class
# =====================================

class Agent:
    def __init__(self, role, goal):
        self.role = role
        self.goal = goal

    def run(self, task):
        prompt = f"""
You are {self.role}.
Goal: {self.goal}

Task:
{task}

Give a clear and structured answer.
"""
        return gemini_call(prompt)

# =====================================
# 🤖 Agents
# =====================================

trend_agent = Agent(
    "Trend Monitoring Agent",
    "Find latest EdTech innovations"
)

summary_agent = Agent(
    "Summarization Agent",
    "Simplify insights"
)

writer_agent = Agent(
    "Blog Writer",
    "Write engaging blog posts"
)

publisher_agent = Agent(
    "Publisher",
    "Format content professionally"
)

# =====================================
# 🔄 Workflow
# =====================================

def run_system():

    print("\n🔍 Step 1: Trends")
    trends = trend_agent.run("Find 5 latest EdTech trends in 2026")
    print(trends)

    print("\n📊 Step 2: Summary")
    summary = summary_agent.run(trends)
    print(summary)

    print("\n✍️ Step 3: Blog")
    blog = writer_agent.run(summary)
    print(blog)

    print("\n🚀 Step 4: Publish")
    final = publisher_agent.run(blog)
    print(final)

    return final

# =====================================
# 🚀 RUN
# =====================================

result = run_system()

print("\n🎉 FINAL OUTPUT:\n", result)
```
