```python
# A company surveyed the salaries of 1,000 employees. The average salary was around $50,000, with most employees earning
#  within a spread of about $10,000.
# The HR department wants to visualize the distribution of salaries to understand pay structure and
#  identify whether salaries are
#  clustered around the average or spread out widely.
# You are asked to:
# - Plot a histogram showing the salary distribution.
# - Add a title to make the chart clear.
# - Interpret the chart:
# - What does the shape of the distribution suggest about how salaries are spread?
# - Are most employees earning close to the average, or is there a wide variation?
# - How could HR use this information to review compensation policies?

import numpy as np
sal=np.random.normal(50000,10000,1000)
plt.hist(sal,bins=20)
plt.show()
```

```python
# 📘 Scenario Question
# A professor recorded the exam scores of 1,000 students in a statistics course. The average 
# score was around 75 marks, with most
# students scoring within a spread of about 10 marks.
# The professor wants to visualize the distribution of exam scores to understand how students
#  performed overall and
#  whether the scores follow a normal pattern.
# You are asked to:
# - Plot a histogram showing the distribution of exam scores.
# - Add a title and axis labels to make the chart clear.
# - Interpret the chart:
# - What does the shape of the distribution suggest about student performance?
# - Are most students clustered around the average, or is there wide variation?
# - How could the professor use this information to adjust teaching methods?

import matplotlib.pyplot as plt
import numpy as np
marks=np.random.normal(75,10,1000)
plt.hist(marks,bins=20)
plt.title("Performance")
plt.xlabel("marks")
plt.ylabel("Students")
plt.show()
```

```python
# 📘 Scenario Question
# A company wants to analyze whether advertising spend impacts sales revenue. The marketing team collected data from
#  several campaigns, where ad spend ranged from $2,000 to $18,000 and the corresponding sales revenue was recorded.
# You are asked to:
# - Plot a scatter chart showing the relationship between ad spend and sales.
# - Add a trend line to highlight the overall pattern.
# - Interpret the chart: does higher ad spend generally lead to higher sales?

# Data : Ad spend vs sales
ad_spend=np.array([100,200,140,25])
sales=np.array([456,321,789,254])
# scatter plot
plt.scatter(ad_spend,sales,color='blue',alpha=0.7)
# Trend Line
z=np.polyfit(ad_spend,sales,1)
p=np.poly1d(z)
plt.plot(ad_spend,p(ad_spend),'r--',label='Trend')
#labels and title
plt.xlabel('Ad Spend ($)')
plt.ylabel('Sales ($)')
plt.title('Ad Spend vs sales')
plt.legend()


```
