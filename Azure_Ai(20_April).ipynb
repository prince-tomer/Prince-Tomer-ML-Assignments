```python
# Scenario
# A student is building a simple AI assistant in Google Colab for a class project. The assistant is
# hosted on Azure Foundry, and the student wants to ask it questions like:

# “What can you help with?”

# “Explain this topic simply.”

# “Give me a short summary.”

# Instead of talking directly to the model through a chat app, the student uses Python code to connect to the Azure service.

from google.colab import userdata
from openai import OpenAI

api_key = userdata.get("azure_api_key")
my_endpoint = "https://res20april.services.ai.azure.com/api/projects/proj-default"
my_agent = "email-assistant"
my_version = "2"


openai_client = OpenAI(
    base_url=f"{my_endpoint}/openai/v1",
    api_key=api_key
)

response = openai_client.responses.create(
    input=[{"role": "user", "content": "Tell me what you can help with."}],
    extra_body={
        "agent_reference": {
            "name": my_agent,
            "version": my_version,
            "type": "agent_reference"
        }
    },
)

print(response.output_text)
```

```python
# Install
!pip install openai

from openai import OpenAI
from google.colab import userdata


api_key = userdata.get('azure_api_key')


my_endpoint = "https://res20april.services.ai.azure.com/api/projects/proj-default"


my_agent = "interview-simulator"
my_version = "1"


client = OpenAI(
    base_url=f"{my_endpoint}/openai/v1",
    api_key=api_key
)


mode = input("Choose mode (HR / Technical / Stress): ")


start = client.responses.create(
    input=[{
        "role": "user",
        "content": f"Start a {mode} interview. Ask one question only."
    }],
    extra_body={
        "agent_reference": {
            "name": my_agent,
            "version": my_version,
            "type": "agent_reference"
        }
    },
)

question = start.output_text
print("\n Interview Question:\n", question)


answer = input("\nYour Answer: ")


feedback = client.responses.create(
    input=[{
        "role": "user",
        "content": f"""
Evaluate this interview answer:

Question: {question}
Answer: {answer}

Give:
- Strengths
- Weaknesses
- Improvements
- Better sample answer
"""
    }],
    extra_body={
        "agent_reference": {
            "name": my_agent,
            "version": my_version,
            "type": "agent_reference"
        }
    },
)

print("\n Feedback:\n", feedback.output_text)
```
