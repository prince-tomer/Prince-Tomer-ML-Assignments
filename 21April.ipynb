```python
import requests
from google.colab import userdata
from azure.storage.blob import BlobServiceClient
import datetime

api_key = userdata.get("Ai-foundry")
blob_conn_str = userdata.get("BLOB_CONNECTION")

url = "https://harshita-ai-foundry.services.ai.azure.com/api/projects/proj-default/applications/resume-analyzer/protocols/openai/responses?api-version=2025-11-15-preview"


headers = {
    "Content-Type": "application/json",
    "api-key": api_key
}

print("Paste resume (press Enter twice to finish):")
lines = []
while True:
    line = input()
    if line == "":
        break
    lines.append(line)

resume_text = "\n".join(lines)

data = {
    "input": [
        {
            "role": "user",
            "content": resume_text
        }
    ]
}

response = requests.post(url, headers=headers, json=data)

print("\nStatus Code:", response.status_code)
print("Raw Response:", response.text)

if response.status_code == 200:
    try:
        result = response.json()
        output = result.get("output_text", str(result))
    except:
        output = response.text
else:
    output = f"Error: {response.status_code} - {response.text}"

print("\nAI ANALYSIS:\n")
print(output)

blob_service_client = BlobServiceClient.from_connection_string(blob_conn_str)

file_name = f"resume_{datetime.datetime.now().strftime('%Y%m%d_%H%M%S')}.txt"

blob_client = blob_service_client.get_blob_client(
    container="resume-output",
    blob=file_name
)

blob_client.upload_blob(output, overwrite=True)

print("\n Saved to Azure Blob Storage as:", file_name)
```

```python

import requests
from google.colab import userdata
from azure.storage.blob import BlobServiceClient
import datetime

api_key = userdata.get("Ai-foundry")
blob_conn_str = userdata.get("BLOB_CONNECTION")

url = "https://harshita-ai-foundry.services.ai.azure.com/api/projects/proj-default/applications/research-summarizer/protocols/openai/responses?api-version=2025-11-15-preview"
headers = {
    "Content-Type": "application/json",
    "api-key": api_key
}

print("Paste research paper text (press Enter twice to finish):")
lines = []
while True:
    line = input()
    if line == "":
        break
    lines.append(line)

paper_text = "\n".join(lines)

data = {
    "input": [
        {
            "role": "user",
            "content": paper_text
        }
    ]
}

response = requests.post(url, headers=headers, json=data)

print("\nStatus Code:", response.status_code)

if response.status_code == 200:
    try:
        result = response.json()
        output = result["output"][0]["content"][0]["text"]
    except Exception as e:
        output = f"Parsing error: {e}"
else:
    output = f"Error: {response.status_code} - {response.text}"

print("\n RESEARCH SUMMARY:\n")
print(output)

blob_service_client = BlobServiceClient.from_connection_string(blob_conn_str)

file_name = f"research_summary_{datetime.datetime.now().strftime('%Y%m%d_%H%M%S')}.txt"

blob_client = blob_service_client.get_blob_client(
    container="resume-output",  # you can reuse same container
    blob=file_name
)

blob_client.upload_blob(output, overwrite=True)

print("\nSaved to Azure Blob Storage as:", file_name)
```
