```python
# ================================
# 🧠 Shared Memory
# ================================
class SharedMemory:
    def __init__(self):
        self.data = {}

    def update(self, agent_name, result):
        print(f"\n[Memory Updated by {agent_name}]")
        self.data[agent_name] = result

    def get(self, agent_name):
        return self.data.get(agent_name, "")


# ================================
# 🔍 Search Agent
# ================================
class SearchAgent:
    name = "search"

    def run(self, goal, memory):
        print("\n[Search Agent] Collecting data...")

        return """
EV market in India is growing rapidly.
Government subsidies boosting demand.
High adoption in metro cities.
"""


# ================================
# 📊 Analyst Agent
# ================================
class AnalystAgent:
    name = "analyst"

    def run(self, goal, memory):
        print("\n[Analyst Agent] Analyzing data...")

        data = memory.get("search")

        return f"""
Key Insights:
- Rapid EV growth in India
- Government policies driving adoption
- Strong demand in urban areas
"""


# ================================
# ✍️ Writer Agent
# ================================
class WriterAgent:
    name = "writer"

    def run(self, goal, memory):
        print("\n[Writer Agent] Writing report...")

        analysis = memory.get("analyst")

        return f"""
📄 EV Market Report (India 2025)

{analysis}

Conclusion:
EV industry has strong future potential.
"""


# ================================
# ✅ QA Agent
# ================================
class QAAgent:
    name = "qa"

    def run(self, goal, memory):
        print("\n[QA Agent] Checking report...")

        report = memory.get("writer")

        return report + "\n\n✅ QA Passed: Report is clear and structured."


# ================================
# 🎼 Orchestrator
# ================================
class Orchestrator:

    def __init__(self):
        self.memory = SharedMemory()
        self.agents = [
            SearchAgent(),
            AnalystAgent(),
            WriterAgent(),
            QAAgent()
        ]

    def run(self, goal):
        print("\n🎯 Goal:", goal)

        for agent in self.agents:
            result = agent.run(goal, self.memory)
            self.memory.update(agent.name, result)

        return self.memory.get("qa")


# ================================
# 🚀 Run System
# ================================
goal = "Write a comprehensive market report on EV industry trends in India for 2025"

system = Orchestrator()
result = system.run(goal)

print("\n📢 FINAL OUTPUT:\n")
print(result)
```
