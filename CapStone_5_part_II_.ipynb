```python

# AI Research Assistant – Working Gemini Code


from google import genai
from transformers import AutoTokenizer

from google.colab import userdata
from google import genai

# Get API key from Colab Secrets
api_key = userdata.get("api_key")
# Replace with your Gemini API key
client = genai.Client(api_key=api_key)


# TASK 1 – LLM Interaction


print("\n            TASK 1 : LLM INTERACTION            \n")

article = """
Artificial Intelligence is transforming modern education. AI-powered tools
enable personalized learning experiences by analyzing student performance
data and adapting teaching methods accordingly. Universities are adopting
AI systems to automate administrative tasks, detect plagiarism, and provide
intelligent tutoring systems. However, concerns regarding data privacy,
ethical usage, and over-reliance on automation remain significant challenges.
Experts believe AI will support educators rather than replace them.
"""

response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents=f"Summarize the following article:\n{article}"
)

print("Summary:\n")
print(response.text)



# TASK 2 – Prompt Engineering


print("\n            TASK 2 : PROMPT ENGINEERING               \n")

zero_prompt = f"""
Summarize the following article in 5 bullet points:

{article}
"""

response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents=zero_prompt
)

print("Zero Shot Output:\n")
print(response.text)



# TASK 3 – Prompt Optimization


print("\n                TASK 3 : PROMPT OPTIMIZATION              \n")

optimized_prompt = f"""
You are an academic research assistant.

Provide:
3 Key Insights
1 Actionable Recommendation

Article:
{article}
"""

response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents=optimized_prompt
)

print(response.text)



# TASK 4 – Tokenization


print("\n                 TASK 4 : TOKENIZATION                  \n")

tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

text = "Artificial intelligence is transforming education systems worldwide."

tokens = tokenizer.tokenize(text)

print("Tokens:", tokens)
print("Token Count:", len(tokens))



# TASK 5 – Mini AI Tool


print("\n              TASK 5 : MINI AI ASSISTANT                  \n")

user_article = input("Enter a research article:\n")

prompt = f"""
Generate:

1. Short Summary
2. Three Key Insights
3. One Actionable Recommendation

Article:
{user_article}
"""

response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents=prompt
)

print("\nAI Assistant Output:\n")
print(response.text)
```
