```python
#  Scenario: University Library Search
# Imagine you’re building a smart search system for a university library.
# - Problem: Students often type queries like “fast dog running” or “canine sprinting”, but
# the library catalog has books titled “The dog runs fast”. Traditional keyword search misses these connections.
# - Solution: Use embeddings to capture semantic meaning instead of exact words.


from sentence_transformers import SentenceTransformer

# Choose from popular models:
# "all-MiniLM-L6-v2" → 384 dims, lightweight, fast
# "all-mpnet-base-v2" → 768 dims, stronger semantic understanding
# "BAAI/bge-m3" → 1024 dims, multilingual + dense/sparse retrieval
# "sentence-transformers/e5-large-v2" → 1024 dims, top performer on MTEB

model_name = "all-MiniLM-L6-v2"   # swap this to test other models
model = SentenceTransformer(model_name)

texts = [
    "The dog runs fast",
    "A quick canine sprints",
    "Python is a programming language"
]

embeddings = model.encode(texts)

print(f"Model: {model_name}")
print(f"Embedding shape: {embeddings.shape}")   # e.g. (3, 384) for MiniLM
print(f"First vector sample: {embeddings[0][:10]}")  # show first 10 values
```

```python
"""Scenario: Hospital Patient Record Search
- Problem: Doctors and nurses often type queries like “heart attack” or “chest pain emergency”, 
but the hospital’s electronic health record (EHR) system stores diagnoses as “acute myocardial infarction”. 
Traditional keyword search fails because medical staff use everyday language while records use clinical terminology.
- Solution: Use embeddings to capture semantic meaning. Embeddings understand that “heart attack” and “acute myocardial infarction” are
related concepts. This allows the search system to return the right patient records even when the query wording doesn’t match the stored terminology."""
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

# Load embedding model
model_name = "sentence-transformers/all-MiniLM-L6-v2"
model = SentenceTransformer(model_name)

# Hospital patient records (clinical terminology)
records = [
    "Patient diagnosed with acute myocardial infarction",
    "Individual suffering from hypertension",
    "Patient shows symptoms of diabetes mellitus",
    "Case of chronic kidney disease"
]

# Doctor or nurse search query (everyday language)
query = "heart attack emergency"

# Convert text to embeddings
record_embeddings = model.encode(records)
query_embedding = model.encode([query])

# Calculate similarity between query and records
similarity_scores = cosine_similarity(query_embedding, record_embeddings)

# Find best matching record
best_match_index = similarity_scores.argmax()

print("Search Query:", query)
print("\nMost Relevant Record:")
print(records[best_match_index])

print("\nSimilarity Scores:")
for record, score in zip(records, similarity_scores[0]):
    print(f"{record} -> {score:.4f}")
```
