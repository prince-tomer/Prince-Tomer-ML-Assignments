```python
import random
head=0
total=1000 # total 
for i in range(total):
    result=random.randint(0,1)
    if result==1:
        head=head+1
tail=total-head
print("P(h):",head/total )
print("P(t):",tail/total) 
        
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 200 random heights 
heights=np.random.normal(165,10,200)

print("Shortest :",round(heights.min(),1))
print("tallest :",round(heights.max(),1))
print("average :",round(heights.mean(),1))
plt.hist(heights,bins=20)
plt.axvline(heights.mean(),color='red')


      
      
```
